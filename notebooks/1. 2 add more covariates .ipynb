```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## mapping file

```python
mf = pd.read_csv('../data/mapping_cleaned_MrOS.txt', sep='\t', dtype=str, index_col='#SampleID') 
```

```python
# add health variable
health = pd.read_csv('../data/MrOS_healthvari.csv', sep=',', dtype=str)
health = health.rename(columns={'ID': '#SampleID'}).set_index('#SampleID')
health['QLCOMP'] = health['QLCOMP'].astype('category')
```

```python
health.head()
```

```python
print(mf.shape)
mf.head()
```

```python
df = pd.merge(mf, health, left_index=True, right_index=True)
```

```python
print(df.shape)
df.head()
```

```python
df.to_csv('../data/mapping_more_MrOS.txt', sep= '\t', index=True)
```
